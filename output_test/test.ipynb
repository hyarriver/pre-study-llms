```python
answers = {
    1: "",  # Q1?...
}
```

```python
# 自动评分（答案已嵌入，仅本地运行）
correct_answers = {'1': 'A'}
explanations = {'1': 'Because'}

score = 0
total = len(correct_answers)
for qid, ans in correct_answers.items():
    if str(answers.get(int(qid), "")).strip().lower() == str(ans).strip().lower():
        score += 1
    else:
        print(f"题 {qid} 错误。解析: {explanations.get(qid, '')}")
print(f"得分: {score}/{total} ({100*score//max(1,total)}%)")
```
